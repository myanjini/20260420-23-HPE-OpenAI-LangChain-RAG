```python
# 패키지 임포트
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
# 환경 변수 읽어와서 설정하고, 메시지를 작성해서 LLM에 질의
load_dotenv()

model = ChatOpenAI(model="gpt-4o-mini")

messages = [
    SystemMessage(content="당신은 미녀와 야수 이야기에 나오는 미녀 역할로 캐릭터 성격에 맞게 대답해야 합니다."),
    HumanMessage(content="안녕하세요. 나는 게스톤입니다. 오늘 저녁 식사 같이 할까요?")    
]

response = model.invoke(messages)
print(response.content)
```

```python
# StrOutputParser를 이용해서 텍스트만 반환
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

result = parser.invoke(response)
print(result)
```

```python
# 모델 실행과 출력 파서를 체인으로 연결해서 실행
chain = model | parser

result = chain.invoke(messages)
print(result)
```

```python
# 템플릿을 정의 
system_template = "당신은 {story} 이야기에 나오는 {character_a} 역할로 캐릭터 성격에 맞게 대답해야 합니다."
human_template = "안녕하세요. 나는 {character_b}입니다. 오늘 {activity} 같이 할까요?"

from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate([("system", system_template), ("human", human_template)])

prompt = prompt_template.invoke({
    "story": "미녀와 야수", 
    "character_a": "미녀", 
    "character_b": "게시톤", 
    "activity": "저녁 식사"
})
print(type(prompt))
print(prompt)
```

```python
# 프롬프트 템플릿 -> 프롬프트 -> LLM 호출 -> 응답 파싱
prompt = prompt_template.invoke({
    "story": "미녀와 야수", 
    "character_a": "미녀", 
    "character_b": "게시톤", 
    "activity": "저녁 식사"
})
response = model.invoke(prompt)
result = parser.invoke(response)
print(result)
```

```python
# 프롬프트 템플릿 | LLM 호출 | 응답 파싱 체인을 정의해서 실행
chain = prompt_template | model | parser
result = chain.invoke({
    "story": "미녀와 야수", 
    "character_a": "미녀", 
    "character_b": "게시톤", 
    "activity": "저녁 식사"
})
print(result)
```
